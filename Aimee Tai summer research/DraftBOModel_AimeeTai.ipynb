```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import RBF, WhiteKernel, ConstantKernel
from sklearn.model_selection import cross_val_score
from skopt import gp_minimize
from skopt.space import Real
from sklearn.preprocessing import StandardScaler

import warnings
warnings.filterwarnings("ignore")

propertyData = pd.read_csv('property_data.csv')
propertyData = propertyData.sort_values(by='Concentration')
```

```python
# CONTROL
goodData = propertyData[propertyData['Fit'] == True]
heatData = goodData[goodData['Heating'] == True]
noheatData = goodData[goodData['Heating'] == False]

# Find best noise level
def objectiveHeat(alpha):
    alpha = alpha[0] 
    kernel = ConstantKernel(1.0, (1e-2, 1e2)) * RBF(1, (1e-2, 1e2))
    gpr = GaussianProcessRegressor(kernel=kernel, n_restarts_optimizer=10, alpha=alpha)
    # Use negative mean cross-validated score as the objective to minimize
    score = -np.mean(cross_val_score(gpr, xHeat, yHeat, cv=3, scoring='neg_mean_squared_error', error_score='raise'))
    return score
def objectiveNoHeat(alpha):
    alpha = alpha[0] 
    kernel = ConstantKernel(1.0, (1e-2, 1e2)) * RBF(1, (1e-2, 1e2))
    gpr = GaussianProcessRegressor(kernel=kernel, n_restarts_optimizer=10, alpha=alpha)
    # Use negative mean cross-validated score as the objective to minimize
    score = -np.mean(cross_val_score(gpr, xNoHeat, yNoHeat, cv=3, scoring='neg_mean_squared_error', error_score='raise'))
    return score

# Data
xHeat = heatData[['Concentration']].values.reshape(-1, 1)
yHeat = heatData['Elastic Modulus']
xNoHeat = noheatData[['Concentration']].values.reshape(-1, 1)
yNoHeat = noheatData['Elastic Modulus']

# Find best noise level
search_space = [Real(1e-10, 1, "log-uniform")]
resultHeat = gp_minimize(objectiveHeat, search_space, n_calls=50, random_state=0)
resultNoHeat = gp_minimize(objectiveNoHeat, search_space, n_calls=50, random_state=0)

# GPR Model
kernel = RBF(1, (1e-2, 1e2)) #+ WhiteKernel(noise_level=1e-12, noise_level_bounds=(1e-15, 1e10))
gpHeat = GaussianProcessRegressor(kernel=kernel, n_restarts_optimizer=10, alpha=resultHeat.x[0])
gpNoHeat = GaussianProcessRegressor(kernel=kernel, n_restarts_optimizer=10, alpha=resultNoHeat.x[0])
gpHeat.fit(xHeat, yHeat)
gpNoHeat.fit(xNoHeat, yNoHeat)

# Prediction
XTest = np.linspace(8, 20, 1000).reshape(-1, 1)
yPredHeat, yPredStdHeat = gpHeat.predict(XTest, return_std=True)
yPredNoHeat, yPredStdNoHeat = gpNoHeat.predict(XTest, return_std=True)

nextConcentrationHeat = XTest[np.argmax(yPredHeat)]
nextConcentrationNoHeat = XTest[np.argmax(yPredNoHeat)]

# Plot the results on the original axis
plt.figure(figsize=(10, 5))
plt.plot(np.linspace(8, 20, 1000), yPredHeat, 'b-', label='GPR Prediction Heating')
plt.fill_between(np.linspace(8, 20, 1000), yPredHeat - 1.96 * yPredStdHeat, yPredHeat + 1.96 * yPredStdHeat, alpha=0.2, color='b')
plt.scatter(xHeat.flatten(), yHeat, label='Heating Samples', color='b')
plt.scatter(xNoHeat.flatten(), yNoHeat, label='No Heating Samples', color='r')
plt.fill_between(np.linspace(8, 20, 1000), yPredNoHeat - 1.96 * yPredStdNoHeat, yPredNoHeat + 1.96 * yPredStdNoHeat, alpha=0.2, color='r')
plt.plot(np.linspace(8, 20, 1000), yPredNoHeat, 'r-', label='GPR Prediction No Heating')
plt.xlabel('Polymer Concentration (wt%)')
plt.ylabel("Young's Modulus")
plt.legend(loc='upper left')
plt.title("CONTROL Gaussian Process Regression for Young's Modulus vs. Polymer Concentration")
plt.show()

print(f"Suggested concentration to test next (Heating): {nextConcentrationHeat[0]}")
print(f"Suggested concentration to test next (No Heating): {nextConcentrationNoHeat[0]}")
```

```python
# NORMALIZED & HYPERPARAMETER TUNING
goodData = propertyData[propertyData['Fit'] == True]
heatData = goodData[goodData['Heating'] == True]
noheatData = goodData[goodData['Heating'] == False]

from sklearn.model_selection import GridSearchCV

# Aggregate mean and variance
aggregatedHeatData = heatData.groupby('Concentration').agg(
    meanYHeat=('Elastic Modulus', 'mean'),
    varYHeat=('Elastic Modulus', 'var'),
    countHeat=('Elastic Modulus', 'count')
).reset_index()

aggregatedNoHeatData = noheatData.groupby('Concentration').agg(
    meanYNoHeat=('Elastic Modulus', 'mean'),
    varYNoHeat=('Elastic Modulus', 'var'),
    countNoHeat=('Elastic Modulus', 'count')
).reset_index()

# Scaler
scalerHeatX = StandardScaler()
scalerNoHeatX = StandardScaler()

scalerHeatY = StandardScaler()
scalerNoHeatY = StandardScaler()

scalerHeatX.fit(aggregatedHeatData[['Concentration']])
scalerNoHeatX.fit(aggregatedNoHeatData[['Concentration']])

scalerHeatY.fit(aggregatedHeatData[['meanYHeat']].values.reshape(-1, 1))
scalerNoHeatY.fit(aggregatedNoHeatData[['meanYNoHeat']].values.reshape(-1, 1))

# Scale the data
X_heat_scaled = scalerHeatX.transform(aggregatedHeatData[['Concentration']])
Y_heat_scaled = scalerHeatY.transform(aggregatedHeatData[['meanYHeat']].values.reshape(-1, 1)).flatten()

X_noheat_scaled = scalerNoHeatX.transform(aggregatedNoHeatData[['Concentration']])
Y_noheat_scaled = scalerNoHeatY.transform(aggregatedNoHeatData[['meanYNoHeat']].values.reshape(-1, 1)).flatten()

# Define kernel
kernel = ConstantKernel(1.0, (1e-2, 1e2)) * RBF(length_scale=1.0, length_scale_bounds=(1e-2, 1e2)) + WhiteKernel(noise_level=1, noise_level_bounds=(1e-10, 1e1))

# Determine the number of splits
n_splits_heat = min(5, len(X_heat_scaled))
n_splits_noheat = min(5, len(X_noheat_scaled))

# Hyperparameter tuning using GridSearchCV
param_grid = {
    "kernel__k1__k1__constant_value": [0.1, 1, 10],
    "kernel__k1__k2__length_scale": [0.1, 1, 10],
    "kernel__k2__noise_level": [0.01, 0.1, 1]
}

gpr = GaussianProcessRegressor(kernel=kernel, n_restarts_optimizer=10)

try:
    grid_search_heat = GridSearchCV(gpr, param_grid, cv=n_splits_heat, n_jobs=-1)
    grid_search_heat.fit(X_heat_scaled, Y_heat_scaled)
    best_gpr_heat = grid_search_heat.best_estimator_
    print("Heat model fitting completed successfully.")
except Exception as e:
    print(f"Error during fitting heat data: {e}")

try:
    grid_search_noheat = GridSearchCV(gpr, param_grid, cv=n_splits_noheat, n_jobs=-1)
    grid_search_noheat.fit(X_noheat_scaled, Y_noheat_scaled)
    best_gpr_noheat = grid_search_noheat.best_estimator_
    print("No heat model fitting completed successfully.")
except Exception as e:
    print(f"Error during fitting no heat data: {e}")

# If fitting was successful, proceed with prediction
if 'best_gpr_heat' in locals() and 'best_gpr_noheat' in locals():
    # Predict
    XTest = np.linspace(8, 22, 1000).reshape(-1, 1)
    XTest_scaled_heat = scalerHeatX.transform(XTest)
    XTest_scaled_noheat = scalerNoHeatX.transform(XTest)

    yPredHeat, yPredStdHeat = best_gpr_heat.predict(XTest_scaled_heat, return_std=True)
    yPredNoHeat, yPredStdNoHeat = best_gpr_noheat.predict(XTest_scaled_noheat, return_std=True)

    # Inverse transform predictions
    yPredHeat = scalerHeatY.inverse_transform(yPredHeat.reshape(-1, 1)).flatten()
    yPredNoHeat = scalerNoHeatY.inverse_transform(yPredNoHeat.reshape(-1, 1)).flatten()
    
    yPredStdHeat = yPredStdHeat * scalerHeatY.scale_
    yPredStdNoHeat = yPredStdNoHeat * scalerNoHeatY.scale_

    # Next concentration
    nextConcentrationHeat = XTest[np.argmax(yPredHeat)]
    nextConcentrationNoHeat = XTest[np.argmax(yPredNoHeat)]

    # Plot results
    plt.figure(figsize=(10, 5))
    plt.errorbar(aggregatedHeatData['Concentration'], aggregatedHeatData['meanYHeat'], yerr=np.sqrt(aggregatedHeatData['varYHeat']), fmt='o', label='Heating Average', color='b')
    plt.plot(XTest, yPredHeat, label='GPR Prediction Heating', color='b')
    plt.fill_between(XTest.ravel(), yPredHeat - 1.96 * yPredStdHeat.flatten(), yPredHeat + 1.96 * yPredStdHeat.flatten(), alpha=0.2, color='b')

    plt.errorbar(aggregatedNoHeatData['Concentration'], aggregatedNoHeatData['meanYNoHeat'], yerr=np.sqrt(aggregatedNoHeatData['varYNoHeat']), fmt='o', label='No Heating Average', color='r')
    plt.plot(XTest, yPredNoHeat, label='GPR Prediction No Heating', color='r')
    plt.fill_between(XTest.ravel(), yPredNoHeat - 1.96 * yPredStdNoHeat.flatten(), yPredNoHeat + 1.96 * yPredStdNoHeat.flatten(), alpha=0.2, color='r')

    plt.xlabel('Polymer Concentration (wt%)')
    plt.ylabel("Young's Modulus")
    plt.legend(loc='upper left')
    plt.title("HYPERPARAMETER TUNING Gaussian Process Regression for Young's Modulus vs. Polymer Concentration")
    plt.show()
else:
    print("Fitting was not successful. Unable to proceed with predictions.")
```
