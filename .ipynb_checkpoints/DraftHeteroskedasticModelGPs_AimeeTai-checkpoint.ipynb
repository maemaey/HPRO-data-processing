```python
#modulus_mean ~ f1(concentration)
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import RBF, WhiteKernel, ConstantKernel
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV

import warnings
warnings.filterwarnings("ignore")


allData = pd.read_csv('combinedData.csv')
allData = allData.sort_values(by='Polymer Concentration Real')
allData = allData[allData['Fit'] == True]

heatData = allData[allData['Heating'] == True]
noheatData = allData[allData['Heating'] == False]

xHeat = heatData['Polymer Concentration Real'].values.reshape(-1, 1)
yHeat = heatData['Elastic Modulus']
xNoHeat = noheatData['Polymer Concentration Real'].values.reshape(-1, 1)
yNoHeat = noheatData['Elastic Modulus']

#Define kernel
kernelHeat = ConstantKernel(1.0, (1e-2, 1e2)) * RBF(length_scale=5, length_scale_bounds=(1e-2, 1e3)) + WhiteKernel(noise_level=5, noise_level_bounds=(1, 50))
kernelNoHeat = ConstantKernel(1.0, (1e-2, 1e2)) * RBF(length_scale=5, length_scale_bounds=(1e-2, 1e3)) + WhiteKernel(noise_level=5, noise_level_bounds=(1, 50))

#Determine the number of splits
n_splits_heat = min(5, len(xHeat))
n_splits_noheat = min(5, len(xNoHeat))

#Hyperparameter tuning using GridSearchCV
param_grid = {
    "kernel__k1__k1__constant_value": np.logspace(0, 2, 5),
    "kernel__k1__k2__length_scale": np.logspace(-2, 3, 5), 
    "kernel__k2__noise_level": np.logspace(0, 1.7, 5)  
}

gprHeat = GaussianProcessRegressor(kernel=kernelHeat, n_restarts_optimizer=10)
gprNoHeat = GaussianProcessRegressor(kernel=kernelNoHeat, n_restarts_optimizer=10)

try:
    grid_search_heat = GridSearchCV(gprHeat, param_grid, cv=n_splits_heat, n_jobs=-1, error_score='raise')
    grid_search_heat.fit(xHeat, yHeat)
    best_gpr_heat = grid_search_heat.best_estimator_
    print("Heat model fitting completed successfully.")
except Exception as e:
    print(f"Error during fitting heat data: {e}")

try:
    grid_search_noheat = GridSearchCV(gprNoHeat, param_grid, cv=n_splits_noheat, n_jobs=-1, error_score='raise')
    grid_search_noheat.fit(xNoHeat, yNoHeat)
    best_gpr_noheat = grid_search_noheat.best_estimator_
    print("No heat model fitting completed successfully.")
except Exception as e:
    print(f"Error during fitting no heat data: {e}")

# If fitting was successful, proceed with prediction
if 'best_gpr_heat' in locals() and 'best_gpr_noheat' in locals():
    # Predict
    XTest = np.linspace(0.08, 0.22, 1000).reshape(-1, 1)

    yPredHeat, yPredStdHeat = best_gpr_heat.predict(XTest, return_std=True)
    yPredNoHeat, yPredStdNoHeat = best_gpr_noheat.predict(XTest, return_std=True)

    # Next concentration
    nextConcentrationHeat = XTest[np.argmax(yPredHeat)]
    nextConcentrationNoHeat = XTest[np.argmax(yPredNoHeat)]

    # Plot results
    plt.figure(figsize=(12, 8))
    plt.scatter(xHeat, yHeat, label='Original Heating Data', color='b')
    plt.plot(XTest, yPredHeat, label='GP Prediction Heating', color='b')
    plt.fill_between(XTest.ravel(), yPredHeat - 1.96 * yPredStdHeat.flatten(), yPredHeat + 1.96 * yPredStdHeat.flatten(), alpha=0.2, color='b')

    plt.scatter(xNoHeat, yNoHeat, label='Original No Heating Data', color='r')
    plt.plot(XTest, yPredNoHeat, label='GPR Prediction No Heating', color='r')
    plt.fill_between(XTest.ravel(), yPredNoHeat - 1.96 * yPredStdNoHeat.flatten(), yPredNoHeat + 1.96 * yPredStdNoHeat.flatten(), alpha=0.2, color='r')

    plt.xlabel('Polymer Concentration')
    plt.ylabel("Young's Modulus (bar)")
    plt.legend(loc='upper left')
    plt.title("modulus_mean ~ f1(concentration)")
    plt.show()
else:
    print("Fitting was not successful. Unable to proceed with predictions.")
```

```python
#modulus_log_variance ~ f2(concentration, humidity, rest_time) - HEAT using sklearn
import pandas as pd
import numpy as np
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import RBF, WhiteKernel, ConstantKernel as C
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score

allData = pd.read_csv('combinedData.csv')
allData = allData.sort_values(by='Polymer Concentration Real')
allData = allData[allData['Fit'] == True]

heatData = allData[allData['Heating'] == True]
noheatData = allData[allData['Heating'] == False]

xHeat = heatData[['Polymer Concentration Real', 'Rest Time (days)', 'Relative Humidity']]
yHeat = heatData['Elastic Modulus']

X_train, X_test, y_train, y_test = train_test_split(xHeat, yHeat, test_size=0.2, random_state=42)

# Define the kernel: ConstantKernel * RBF
kernel = C() * RBF() + WhiteKernel()

# Initialize the Gaussian Process Regressor
gp = GaussianProcessRegressor(kernel=kernel, n_restarts_optimizer=10)

# Fit the model
gp.fit(X_train, y_train)

# Make predictions
y_pred, y_pred_std = gp.predict(X_test, return_std=True)

# Evaluate performance
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f'Mean Squared Error: {mse}')
print(f'R-squared: {r2}')

# Optional: Print kernel details
print(f'Kernel: {gp.kernel_}')
```

```python
#modulus_log_variance ~ f2(concentration, humidity, rest_time) HEAT
import pandas as pd
import torch
import gpytorch
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score

allData = pd.read_csv('combinedData.csv')
allData = allData.sort_values(by='Polymer Concentration Real')
allData = allData[allData['Fit'] == True]

heatData = allData[allData['Heating'] == True]
noheatData = allData[allData['Heating'] == False]

xHeat = heatData[['Polymer Concentration Real', 'Rest Time (days)', 'Relative Humidity']]
yHeat = heatData['Elastic Modulus']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(xHeat, yHeat, test_size=0.2, random_state=42)

# Convert data to PyTorch tensors
X_train_tensor = torch.tensor(X_train.values, dtype=torch.float32)
y_train_tensor = torch.tensor(y_train.values, dtype=torch.float32)
X_test_tensor = torch.tensor(X_test.values, dtype=torch.float32)

# Define the Gaussian Process model
class GPModel(gpytorch.models.ExactGP):
    def __init__(self, train_x, train_y, likelihood):
        super(GPModel, self).__init__(train_x, train_y, likelihood)
        self.mean_module = gpytorch.means.ConstantMean()
        self.covar_module = gpytorch.kernels.RBFKernel()  # Radial Basis Function kernel
        
    def forward(self, x):
        mean_x = self.mean_module(x)
        cov_x = self.covar_module(x)
        return gpytorch.distributions.MultivariateNormal(mean_x, cov_x)

# Define the likelihood
likelihood = gpytorch.likelihoods.GaussianLikelihood()

# Create the model and likelihood
model = GPModel(X_train_tensor, y_train_tensor, likelihood)

# Set up the optimizer and loss function
optimizer = torch.optim.Adam(model.parameters(), lr=0.1)
mll = gpytorch.mlls.ExactMarginalLogLikelihood(likelihood, model)

# Training loop
model.train()
likelihood.train()

for epoch in range(50):  # Number of epochs
    optimizer.zero_grad()
    output = model(X_train_tensor)
    loss = -mll(output, y_train_tensor)
    loss.backward()
    optimizer.step()

# Make predictions
model.eval()
likelihood.eval()

with torch.no_grad():
    test_preds = model(X_test_tensor)
    y_pred = test_preds.mean.numpy()
    y_pred_std = test_preds.stddev.numpy()

# Evaluate performance
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f'Mean Squared Error: {mse}')
print(f'R-squared: {r2}')
```
